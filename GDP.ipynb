```python
import pandas as pd
```

```python
import requests as req

url = "https://countrycode.org"

res = req.get(url)
res.text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
lst_elements = soup.select("tr td a")
countries =[]

for e in lst_elements:
    countries.append(e.text)
    
countries=countries[:240]
print(countries)
```

```python
lst_gdp = soup.select("tr td")


gdp = list(map(lambda x : x.text,lst_gdp))[5::6][:240]
```

```python
df = pd.DataFrame()

df["countries"]=countries
df["gdp"]=gdp
```

```python
lst2 = []

for i in df['gdp']:
    lst2.append(i.split(" "))
```

```python
numeros=[]
for e in lst2:
    if len(e)==1: 
        numeros.append(0)
        
    else: 
        num=float(e[0])
        if len(e)==2:
            if 'M' in e[1]: num*=10e5
            elif 'B' in e[1]: num*=10e8
            elif 'T' in e[1]: num*=10e11
            numeros.append(num)
```

```python
df["gdp"]=numeros
df
```

```python
df.sort_values("gdp", ascending=False)
```

```python
df_cero = df[(df["gdp"]==0)]
display(df_cero)
```

```python
data_final = df.drop(df_cero.index, axis=0)
display(data_final)
```
